```python
import os
import matplotlib
import matplotlib.pyplot as plt
import pandas  as pd
%matplotlib inline
```

```python
root = os.getcwd()
file = os.path.join(root, '../company_sales_data.csv')

data = pd.read_csv(file)
data.head()
```

### 1. Прочитайте все данные о продажах продукта и покажите их с помощью многострочного графика

```python
line_styles = ['-', '--', '-.', ':']
markers = ['o', 'v', '*', 'P', 'x', 'd']
months = data['month_number'].tolist()
products = data.columns.to_list()[1:-2]
for i, prod in enumerate(products):
    values = data[prod].tolist()
    plt.plot(months, values, label=prod, linestyle = line_styles[i % len(line_styles)], linewidth=2, marker = markers[i % len(markers)])
plt.legend(loc='best', bbox_to_anchor=(0.95, 0.9))
plt.grid(True)
plt.xticks(months)
plt.title('Данные о продажах товаров', fontsize = 14, fontweight='bold')
plt.xlabel('Месяц', fontsize = 13)
plt.ylabel('Всего продано', fontsize = 13)
plt.show()
```

### 2. Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью столбчатой диаграммы.

```python
baths = data['bathingsoap'].tolist()
plt.bar(months, baths, label= 'bathingsoap')
plt.xticks(months)
plt.grid(True)
plt.legend(loc='best')
plt.title('Данные о продажах мыла для купания', fontsize = 14, fontweight='bold')
plt.xlabel('Месяц', fontsize = 13)
plt.ylabel('Всего продано', fontsize = 13)
plt.show()
```

### 3. Рассчитайте общие данные о продажах за прошлый год для каждого продукта и покажите их с помощью круговой диаграммы

```python
total = [sum(data[prod].tolist()) for prod in products]

explode = [0.08 for _ in range(len(products))]
plt.title('Общие дынные о продажах за год', fontsize = 14, fontweight='bold', y=1.2)
plt.pie(total, labels  = products, autopct='%1.1f%%', shadow=True, explode = explode, radius=1.5)

plt.show()
```
